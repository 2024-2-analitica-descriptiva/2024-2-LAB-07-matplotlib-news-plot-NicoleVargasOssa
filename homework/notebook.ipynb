```python
import pandas as pd
df = pd.read_csv("../files/input/news.csv",index_col=0)
df
```

```python
import matplotlib.pyplot as plt
import os

#grafica vacia
plt.figure()

colors = {
        'Television': 'dimgray',
        'Newspaper': 'grey',
        'Internet': 'tab:blue',
        'Radio': 'lightgrey'
    }

zorder= {
        'Television': 1,
        'Newspaper': 1,
        'Internet': 2,
        'Radio': 1
    }

linewidths = {
        'Television': 2,
        'Newspaper': 2,
        'Internet': 4,
        'Radio': 2
    }

#la figura itere sobre las columnas del DataFrame
for col in df.columns:
    plt.plot(df[col],
        color= colors[col],
        label=col,
        zorder=zorder[col],
        linewidth = linewidths[col]
)

#titulo 
plt.title("How people get their news", fontsize=16)
#eliminar lineas
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
#eje y invisible
plt.gca().axes.get_yaxis().set_visible(False)

#mostrar valor de arranque y final
for col in df.columns:
    first_year = df.index[0]
    plt.scatter(
        x=first_year,
        y = df[col][first_year],
        color=colors[col],
        zorder=zorder[col],
    )
#introducir el texto con el valor
    plt.text(
        first_year - 0.2,
        df[col][first_year],
        col + " " + str(df[col][first_year]) + "%",
        ha='right',
        va='center',
        color = colors[col]
        )
#mostrar valor de arranque y final
    last_year = df.index[-1]
    plt.scatter(
        x=last_year,
        y = df[col][last_year],
        color=colors[col],
        zorder=zorder[col],
    )

    plt.text(
        first_year + 0.2,
        df[col][last_year],
        col + " " + str(df[col][last_year]) + "%",
        ha='left',
        va='center',
        color = colors[col]
        )

#los puntos que aparecen en elje x, mostrar los años completos
plt.xticks(
    ticks=df.index,
    labels=df.index,
    ha='center'
)

#salvar la figura como png
plt.tight_layout()
#ruta de salvar
os.makedirs("../files/plots/', exist_ok=True")
plt.savefig("../files/plots/news.png")
plt.show()

#agregar leyenda
#plt.legend(loc="upper right")






```
